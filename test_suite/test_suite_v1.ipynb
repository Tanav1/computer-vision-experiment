```python

# Evaluate model per dataset
def evaluate_model(model, dataset_name, all_batches):
    results = []
    print(f"Evaluating {model.name} on {dataset_name}...")
    for batch_number, batch in enumerate(all_batches, start=1):  # Iterate over all batches
        for X_batch, y_batch in batch:  # Unpack the batch
            loss, accuracy = model.evaluate(X_batch, y_batch, verbose=0)
            print(f"  Batch {batch_number}: Loss = {loss:.4f}, Accuracy = {accuracy:.4f}")
            # Save results including batch number
            results.append({
                'Model': model.name,
                'Dataset': dataset_name,
                'Batch Number': batch_number,
                'Loss': loss,
                'Accuracy': accuracy
            })
            break
    return results

# Load batches
all_noise_batches = noise.datasets
all_affine_batches = affine.datasets
all_rotation_batches = center_rotation.datasets

# Store results
all_results = []


models_and_names = [
    (cnn_A, "CNN_A"),
    (cnn_A_baseline, "CNN_A_Baseline"),
    (cnn_B, "CNN_B"),
    (cnn_B_baseline, "CNN_B_Baseline"),
    (cnn_C, "CNN_C"),
    (cnn_C_baseline, "CNN_C_Baseline"),
    (cnn_D, "CNN_D"),
    (cnn_D_baseline, "CNN_D_Baseline"),
]

# Evaluate each model on each dataset
for model, model_name in models_and_names:
    print(f"\nStarting evaluations for {model_name}...\n")

    # Noise Dataset
    results = evaluate_model(model, "Noise Dataset", all_noise_batches)
    all_results.extend(results)

    # Affine Dataset
    results = evaluate_model(model, "Affine Dataset", all_affine_batches)
    all_results.extend(results)

    # Center Rotation Dataset
    results = evaluate_model(model, "Center Rotation Dataset", all_rotation_batches)
    all_results.extend(results)

print("\nEvaluation completed for all models and datasets.\n")

```
